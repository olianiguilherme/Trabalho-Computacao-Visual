```python
#NÃ£o consegui da forma experada, mas segue as tentativas.
import cv2
import numpy as np
import os

path = r'C:\Users\guilh\Desktop\Trabalho Computacao Visual\img_original\corvo.bmp'
path2 = r'C:\Users\guilh\Desktop\Trabalho Computacao Visual\img_original\background.bmp'
path3 = r'C:\Users\guilh\Desktop\Trabalho Computacao Visual\img_original\corvos.bmp'
path4 = r'C:\Users\guilh\Desktop\Trabalho Computacao Visual\img_original\formas.bmp'
path5 = r'C:\Users\guilh\Desktop\Trabalho Computacao Visual\img_original\rainha.bmp'


img1 = cv2.imread(path)
img2 = cv2.imread(path3)
img3 = cv2.imread(path4)
img4 = cv2.imread(path5)

background = cv2.imread(path2)

Min = np.array([0, 157, 0])
Max = np.array([175, 255, 175])

#Lista imgs
imagens = {'Corvo': img1, 'Corvos': img2, 'Formas': img3, 'Rainha': img4}

for name, img in imagens.items():
    mascara = cv2.inRange(img, Min, Max)

    mascara_img = cv2.bitwise_and(img, img, mask=mascara)

     # Canny
    gray = cv2.cvtColor(mascara_img, cv2.COLOR_BGR2GRAY)
    Canny = cv2.Canny(mascara_img, 50, 100)

    #Sobel
    mascara_x = cv2.Sobel(mascara_img, cv2.CV_8U, 1, 0, ksize=5)
    mascara_y = cv2.Sobel(mascara_img, cv2.CV_8U, 0, 1, ksize=5)
    mascara_sobel = cv2.bitwise_or(mascara_x, mascara_y)

    resized = cv2.resize(background, (img.shape[1], img.shape[0]))

    bk_masked = cv2.bitwise_and(resized, resized, mask=mascara)

    mascara_i = cv2.bitwise_and(img, img, mask=cv2.bitwise_not(mascara))

   #Laplaciano
    lapla = cv2.Laplacian(mascara_i, cv2.CV_8U)

    # Gaussiano
    mascara_gauss = cv2.GaussianBlur(mascara_i, (7, 7), 0)

    bk_masked_gauss = cv2.GaussianBlur(bk_masked, (7, 7), 0)

    result = cv2.add(mascara_i, bk_masked)

    result_canny = cv2.bitwise_and(result, result, mask=Canny)

    cv2.imshow(f'Resultado Final - {name}', result)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    #Salvando
    if not os.path.exists('Sobel'):
       os.makedirs('Sobel')
    cv2.imwrite('Sobel/{}_Sobel.bmp'.format(name), result)
    if not os.path.exists('gauss'):
       os.makedirs('gauss')
    cv2.imwrite('gauss/{}_gauss.bmp'.format(name), mascara_gauss)
    if not os.path.exists('Laplaciano'):
       os.makedirs('Laplaciano')
    cv2.imwrite('Laplaciano/{}_Laplaciano.bmp'.format(name), lapla)
    if not os.path.exists('Canny'):
       os.makedirs('Canny')
    cv2.imwrite('Canny/{}_Canny.bmp'.format(name), Canny)

    




```
